```python
class Bird:
    satiety = 'hungry'
    eggs =  'dont have eggs'
    def __init__(self, name, weight):
        self.name = name
        self.weight = weight
    def feed(self):
        self.satiety = 'animal ate food'
    def pick_eggs(self):
        self.eggs = 'we picked eggs'
         

        
        
class Livestock:
    satiety = 'hungry'
    def __init__(self, name, weight):
        self.name = name
        self.weight = weight
    def feed(self):
        self.satiety = 'animal ate food'
    


class Goose(Bird):
    voice = 'krya-krya'
    def __init__(self, name, weight):
        super().__init__(name, weight)
        

class Chicken(Bird):
    voice = 'kudakh-kudakh'
    def __init__(self, name, weight):
        super().__init__(name, weight)
    
class Duck(Bird):
    voice = 'krya-krya'
    def __init__(self, name, weight):
        super().__init__(name, weight)

class Cow(Livestock):
    voice = 'muuuu'
    milk = 'we need milk'
    def __init__(self, name, weight):
        super().__init__(name, weight)
    def get_milk(self):
        self.milk = 'milk received'
class Sheep(Livestock):
    voice = 'beeee'
    wool = 'we dont have'
    def __init__(self, name, weight):
        super().__init__(name, weight)
    def cut(self):
        self.wool = 'we cut the sheep'

class Goat(Livestock):
    voice = 'meeee'
    milk = 'we need milk'
    def __init__(self, name, weight):
        super().__init__(name, weight)
    def get_milk(self):
        self.milk = 'milk received'

    
    
    
goose_1 = Goose('Gray', 3)
goose_2 = Goose('White', 3.5)
cow = Cow('Manka', 500)
sheep_1 = Sheep('Barashek', 50)
sheep_2 = Sheep('Kudryavyy', 65)
chicken_1 = Chicken('KoKo', 1)
chicken_2 = Chicken('Kukareku', 0.8)
goat_1 = Goat('Roga', 41)
goat_2 = Goat('Kopyta', 45)
duck = Duck('Kryakva', 2)


all_animal = [goose_1.name, goose_1.weight, goose_2.name, goose_2.weight, cow.name, cow.weight, sheep_1.name, sheep_1.weight, sheep_2.name, sheep_2.weight, chicken_1.name, chicken_1.weight, chicken_2.name, chicken_2.weight, goat_1.name, goat_1.weight, duck.name, duck.weight]
all_animal_dict = dict(zip(all_animal[::2], all_animal[1::2]))
max_weight = max(all_animal_dict.values())
name_max_weight = max(all_animal_dict.values())
def name_max_weight():
    for key,value in all_animal_dict.items():
        if value == max(all_animal_dict.values()):
            print(key) 
print(max_weight)
print(name_max_weight())
```
